# Elaborando modelo de Machine Learning

```python
from sklearn.linear_model import Ridge
#construção do modelo
reg = Ridge(alpha=.1)
```

O método de erro abaixo:

Erro quadrado médio:

Nsum((|valor_real - valor previsto|) / N)

```python
from sklearn.metrics import mean_squared_error 

mean_squared_error(test["vel_target"], predictions) #obtendo o erro em relação ao target
#Excelente acertividade do modelo
```

## Otimização do sentido do vento

Abaixo temos evidência de que a direção angular dos ventos aumenta proporcionalmente com a velocidade relativa do vento em média até a faixa de 270 graus. 

```python

def analise_angular(nome_col_direc) :
  import statistics
  lis=[]
  lis2=[]
  lis3=[]
  lis4=[]
  count=0
  count2=0
  count3=0
  count4=0
  for i,j in zip(model1[nome_col_direc],model1['vel_target']):
    if 125<i < 224 :
      lis.append(j)
      count+= 1
    if  225<i<315:
      lis2.append(j)
      count2+= 1
    if 316< i or i<45:
      lis3.append(j)
      count3+=1
    if 46< i < 135:
      lis4.append(j)
      count4+=1
  med=sum(lis)/count
  med2=sum(lis2)/count2
  med3=sum(lis3)/count3
  med4=sum(lis4)/count4
  des= statistics.pstdev(lis)
  des2= statistics.pstdev(lis2)
  des3= statistics.pstdev(lis3)
  des4= statistics.pstdev(lis4)
  resp=print('A média de ângulo ~180 é {:.2f}, des padrão: {:.2f} \n a média de ângulo ~270 é {:.2f}, des padrão: {:.2f} \n a média de ângulo ~0 é {:.2f}, des padrão: {:.2f}  \n a média de ângulo ~90 é {:.2f}, des padrão: {:.2f}'.format(med,des,med2,des2,med3,des3,med4,des4) + 30*'-'+'Fim da função')
  return resp

analise_angular('direc_10m') # para a direção a 10 metros de altura

analise_angular('direc_100m') # para direção a 100 metros de altura
```

Portanto, Podemos prosseguir transformando ângulos maiores que a faixa estipulada dos 270 graus (que possuem uma velocidade do vento gradativamente menor) em seus equivalentes trigonométricos do primeiro quadrante considerando o eixo x. Desta forma, se espera aproximar a correlação da velocidade do vento com o ângulo da velocidade do vento, melhorando assim o modelo.

```python

lista= list(range(315,361))
lista2= list(range(45,-1,-1))
lista3=[]
lista4=[]

  
for i in model1['direc_10m']:
  x= round(i)

  if x >= 315:
    for z,t in zip(lista,lista2):
      if x == z: 
        lista3.append(t) 
  else:
    lista3.append(x)


for j in model1['direc_100m']:
  x= round(j)

  if x >= 315:
    for z,t in zip(lista,lista2):
      if x == z: 
        lista4.append(t) 
  else:
    lista4.append(x)
        
model1['direc_10m']= lista3
model1['direc_100m']= lista4

model1.describe()

```
